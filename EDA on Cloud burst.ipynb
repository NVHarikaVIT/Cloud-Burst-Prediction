```python
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
df['Date'].head()
```

```python
numCols = ['MinimumTemperature', 'MaximumTemperature','Rainfall', 'Sunshine','Evaporation','WindGustSpeed','WindSpeed9am',
       'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am',
       'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temperature9am',
       'Temperature3pm']
catCols = ['Date', 'Location','WindGustDirection','WindDirection9am', 'WindDirection3pm','CloudBurst Today', 'CloudBurstTomorrow']
```

```python
for c in catCols:
    df[c] = df[c].fillna("Unknown")

    print(df[c].value_counts())
```

```python
df.isnull().sum()
```

```python
df1 = df.drop(['Date', 'Location','WindGustDirection','WindDirection9am', 'WindDirection3pm'], axis=1)
df1.shape
```

```python
df1.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df1['CloudBurst Today'] = le.fit_transform(df1['CloudBurst Today'])
df1['CloudBurstTomorrow'] = le.fit_transform(df1['CloudBurstTomorrow'])

# no - 0, Unknown - 1, yes - 2
le.classes_, le.fit_transform(le.classes_)
```

```python
df1.corr()
```

```python
df1.corr()['CloudBurst Today'].sort_values(ascending=False)
```

```python
df1.corr()['CloudBurstTomorrow'].sort_values(ascending=False)
```

```python
from sklearn.impute import KNNImputer
knn_impute = KNNImputer(n_neighbors=5)

impute_values = knn_impute.fit_transform(df[numCols])
impute_df = pd.DataFrame(impute_values, columns=numCols)
df[numCols] = impute_df
```

```python
impute_values = knn_impute.fit_transform(df1[numCols])
impute_df = pd.DataFrame(impute_values, columns=numCols)
df1[numCols] = impute_df
```

```python
df1.isnull().sum().sum(), df.isnull().sum().sum()
```

```python
df.shape
```

```python
for col in numCols:
        q1, q2, q3 = np.percentile(df[col], [25,50,75])
        # interquartile range
        iqr = q3-q1
        # limits of outliers
        ll = q1 - 1.5*iqr
        ul = q3 + 1.5*iqr
        df.drop(df[df[col] < ll].index, inplace=True)
        df.drop(df[df[col] > ul].index, inplace=True)
```

```python
for col in numCols:
        q1, q2, q3 = np.percentile(df1[col], [25,50,75])
        # interquartile range
        iqr = q3-q1
        # limits of outliers
        ll = q1 - 1.5*iqr
        ul = q3 + 1.5*iqr
        df1.drop(df1[df1[col] < ll].index, inplace=True)
        df1.drop(df1[df1[col] > ul].index, inplace=True)
```
